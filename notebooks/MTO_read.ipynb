```python
import pandas as pd
import sqlite3
```

```python
# Load the spreadsheet
file_path = '..//data/AMTO_Import_3327_2022-12-06T00_46_31.291870 (1) (3).xlsx'
xls = pd.ExcelFile(file_path)

# Connect to a new SQLite database (this will create the file if it doesn't exist)
db_connection = sqlite3.connect('../data/IR_data.db')
cursor = db_connection.cursor()

# Iterate over each sheet in the workbook
for sheet_name in xls.sheet_names:
    # Read the sheet into a DataFrame
    df = pd.read_excel(xls, sheet_name=sheet_name)
    
    # Remove any spaces and special characters from sheet names for table naming
    table_name = sheet_name.replace(" ", "_").replace("-", "_").replace("(", "").replace(")", "")
    
    # Write the DataFrame to a SQLite table
    df.to_sql(table_name, db_connection, if_exists='replace', index=False)

# Commit changes and close the connection
db_connection.commit()
db_connection.close()
```
